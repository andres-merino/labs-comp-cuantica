```python
import math
import cmath
import numpy as np
import sympy as sp
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex
from IPython.display import display, Markdown
```

## <font color='3facc9'> Definicion de matrices </font>

Tomando el enunciado del PDF, se usan las matrices:

$$
A = \begin{bmatrix}
3 + 2i & 1 - i & 4 - 3i \\
2 + i & -5 + 4i & 6 \\
-3 + 2i & 7 - i & 8 + 3i
\end{bmatrix}, \quad
B = \begin{bmatrix}
2 + 3i & 4 & 1 + 2i \\
3 - 2i & 5 + i & 7 - 3i \\
6 + 4i & -2 & 9 + 5i
\end{bmatrix}
$$

$$
C = \begin{bmatrix}
1 + 4i & 2 - 2i & 3 + i & -5 \\
-2 + 3i & 6 & 4 + 7i & 2 + i \\
-1 + 2i & -6 + 5i & -8i & 7 + 4i \\
-3 - 4i & 3 - i & -5 - 3i & 5i
\end{bmatrix}, \quad
D = \begin{bmatrix}
i & 2 \\
2 - 3i & 4 + 5i
\end{bmatrix}
$$

```python
A = np.array([
    [3 + 2j, 1 - 1j, 4 - 3j],
    [2 + 1j, -5 + 4j, 6 + 0j],
    [-3 + 2j, 7 - 1j, 8 + 3j]
], dtype=complex)

B = np.array([
    [2 + 3j, 4 + 0j, 1 + 2j],
    [3 - 2j, 5 + 1j, 7 - 3j],
    [6 + 4j, -2 + 0j, 9 + 5j]
], dtype=complex)

C = np.array([
    [1 + 4j, 2 - 2j, 3 + 1j, -5 + 0j],
    [-2 + 3j, 6 + 0j, 4 + 7j, 2 + 1j],
    [-1 + 2j, -6 + 5j, -8j, 7 + 4j],
    [-3 - 4j, 3 - 1j, -5 - 3j, 5j]
], dtype=complex)

D = np.array([
    [1j, 2 + 0j],
    [2 - 3j, 4 + 5j]
], dtype=complex)

print("Matriz A:")
display(sp.Matrix(A).applyfunc(sp.nsimplify))
print("\nMatriz B:")
display(sp.Matrix(B).applyfunc(sp.nsimplify))
print("\nMatriz C:")
display(sp.Matrix(C).applyfunc(sp.nsimplify))
print("\nMatriz D:")
display(sp.Matrix(D).applyfunc(sp.nsimplify))
```

## <font color='3facc9'> Ejercicios 1 </font>

Utiliza la libreria de NumPy, junto con los modulos `cmath` y `math`, y obten:

- La forma polar de cada uno de los determinantes y de la traza de cada matriz.

```python
det_A = np.linalg.det(A)
tr_A = np.trace(A)
det_B = np.linalg.det(B)
tr_B = np.trace(B)
det_C = np.linalg.det(C)
tr_C = np.trace(C)
det_D = np.linalg.det(D)
tr_D = np.trace(D)

print("Resultados:")
for nom, det_M, tr_M in [("A", det_A, tr_A), ("B", det_B, tr_B), ("C", det_C, tr_C), ("D", det_D, tr_D)]:
    print(f"det({nom}) = {det_M:.0f}, \ttr({nom}) = {tr_M:.0f}")

print("\nEn forma polar:")
for nom, det_M, tr_M in [("A", det_A, tr_A), ("B", det_B, tr_B), ("C", det_C, tr_C), ("D", det_D, tr_D)]:
    r_det_M = abs(det_M)
    theta_det_M = cmath.phase(det_M)
    r_tr_M = abs(tr_M)
    theta_tr_M = cmath.phase(tr_M)
    print(f"det({nom}) = {r_det_M:.2f} (cos({theta_det_M:.2f}) + i*sin({theta_det_M:.2f})), \ttr({nom}) = {r_tr_M:.2f} (cos({theta_tr_M:.2f}) + i*sin({theta_tr_M:.2f}))")
```

- $2i(\det(D) + \mathrm{tr}(C))AB - (1 + i)\det(C)\mathrm{tr}(D)BA$.

```python
expr = 2j * (det_D + tr_C) * (A @ B) - (1 + 1j) * det_C * tr_D * (B @ A)
print("Resultado:\n")
# Redondeado a 0 decimales
expr_rounded = np.round(expr, decimals=0)
display(sp.Matrix(expr_rounded).applyfunc(sp.nsimplify))
```

- La inversa de cada matriz, si existe.

```python
print("\nInversas:")
for nombre, M in [("A", A), ("B", B), ("C", C), ("D", D)]:
    det_M = np.linalg.det(M)
    if det_M == 0.0:
        print(f"{nombre}: no tiene inversa.")
    else:
        inv_M = np.linalg.inv(M)
        inv_M_rounded = np.round(inv_M, decimals=2)
        print(f"\n{nombre}^-1 =")
        display(sp.Matrix(inv_M_rounded))
```

- Además, usando la definición y, al mismo tiempo, las propiedades de los valores propios, clasifica cada una de las matrices en normales, hermitianas o unitarias.

Para clasificar una matriz $M$ se verifica numericamente:

- Normal: $MM^\dagger = M^\dagger M$.
- Hermitiana: $M = M^\dagger$.
- Unitaria: $M^\dagger M = I$.

Ademas, como apoyo con autovalores:

- Si es hermitiana, los autovalores deben ser reales.
- Si es unitaria, los modulos de los autovalores deben ser 1.

```python
for nombre, M in [("A", A), ("B", B), ("C", C), ("D", D)]:
    M_dag = M.conj().T
    identidad = np.eye(M.shape[0], dtype=complex)
    es_normal = np.allclose(M @ M_dag, M_dag @ M, atol=1e-8)
    es_hermitiana = np.allclose(M, M_dag, atol=1e-8)
    es_unitaria = np.allclose(M_dag @ M, identidad, atol=1e-8)

    autovalores = np.linalg.eigvals(M)
    autovalores_reales = np.allclose(np.imag(autovalores), 0, atol=1e-8)
    modulos_unidad = np.allclose(np.abs(autovalores), 1, atol=1e-8)

    print(f"\nClasificacion de {nombre}:")
    print(f"- Normal: {es_normal};\t Hermitiana: {es_hermitiana};\t Unitaria: {es_unitaria}")
    print(f"\n- Autovalores: {autovalores}")
    print(f"- Autovalores reales (criterio hermitiano): {autovalores_reales}")
    print(f"- |autovalores| = 1 (criterio unitario): {modulos_unidad}")
```

## <font color='3facc9'> Ejercicios 2 </font>

Utiliza NumPy y Qiskit y obten:

- Crear los vectores estado de $\mathbb{C}^2$: $|0\rangle$ y $|1\rangle$.

```python
ket0 = Statevector(np.array([1, 0], dtype=complex))
ket1 = Statevector(np.array([0, 1], dtype=complex))

for ket in [ket0, ket1]:
    # display(ket.draw("latex"))
    label = ket.draw(output="latex_source")
    display(array_to_latex(ket.data, prefix=f"{label} = "))
    latex_code = label + "=" + array_to_latex(ket.data, source=True)
    print(latex_code)
```

- A partir de los vectores estado creados, hallar $|+\rangle$, $|-\rangle$, $|i+\rangle$, $|i-\rangle$ y los estados de Bell, en formato LaTeX.

```python
ket_plus = (ket0 + ket1) / np.sqrt(2)
ket_minus = (ket0 - ket1) / np.sqrt(2)
ket_i_plus = (ket0 + 1j * ket1) / np.sqrt(2)
ket_i_minus = (ket0 - 1j * ket1) / np.sqrt(2)

phi_plus = (np.kron(ket0.data, ket0.data) + np.kron(ket1.data, ket1.data)) / np.sqrt(2)
phi_minus = (np.kron(ket0.data, ket1.data) - np.kron(ket1.data, ket0.data)) / np.sqrt(2)
psi_plus = (np.kron(ket0.data, ket1.data) + np.kron(ket1.data, ket0.data)) / np.sqrt(2)
psi_minus = (np.kron(ket0.data, ket1.data) - np.kron(ket1.data, ket0.data)) / np.sqrt(2)

for label, ket in [
    (r"|+\rangle", ket_plus),
    (r"|-\rangle", ket_minus),
    (r"|i+\rangle", ket_i_plus),
    (r"|i-\rangle", ket_i_minus),
    (r"|\Phi^+\rangle", phi_plus),
    (r"|\Phi^-\rangle", phi_minus),
    (r"|\Psi^+\rangle", psi_plus),
    (r"|\Psi^-\rangle", psi_minus),
]:
    display(array_to_latex(ket.data, prefix=f"{label} = "))
    latex_code = label + "=" + array_to_latex(ket.data, source=True)
    print(latex_code)
```

- A partir de los vectores estado creados, encontrar la base computacional de $\mathbb{C}^4$ (por ejemplo $|10\rangle$), en formato LaTeX.

```python
ket00 = np.kron(ket0.data, ket0.data)
ket01 = np.kron(ket0.data, ket1.data)
ket10 = np.kron(ket1.data, ket0.data)
ket11 = np.kron(ket1.data, ket1.data)

base_00 = Statevector(ket00)
base_01 = Statevector(ket01)
base_10 = Statevector(ket10)
base_11 = Statevector(ket11)

for label, ket in [
    (r"|00\rangle", base_00),
    (r"|01\rangle", base_01),
    (r"|10\rangle", base_10),
    (r"|11\rangle", base_11),
]:
    display(array_to_latex(ket.data, prefix=f"{label} = "))
    latex_code = label + "=" + array_to_latex(ket.data, source=True)
    print(latex_code)
```

- A partir de los vectores estado creados, hallar los vectores $\frac{1}{\sqrt{2}}|000\rangle + \frac{1}{\sqrt{2}}|111\rangle$, $\frac{i}{\sqrt{2}}|000\rangle - \frac{1}{\sqrt{2}}|111\rangle$ y $\frac{1}{\sqrt{2}}|000\rangle + \frac{i}{\sqrt{2}}|111\rangle$, en formato LaTeX.

```python
ket000 = np.kron(np.kron(ket0.data, ket0.data), ket0.data)
ket111 = np.kron(np.kron(ket1.data, ket1.data), ket1.data)

v1 = Statevector((ket000 + ket111) / np.sqrt(2))
v2 = Statevector((1j * ket000 - ket111) / np.sqrt(2))
v3 = Statevector((ket000 + 1j * ket111) / np.sqrt(2))

for label, ket in [
    (r"\frac{1}{\sqrt{2}}|000\rangle + \frac{1}{\sqrt{2}}|111\rangle", v1),
    (r"\frac{i}{\sqrt{2}}|000\rangle - \frac{1}{\sqrt{2}}|111\rangle", v2),
    (r"\frac{1}{\sqrt{2}}|000\rangle + \frac{i}{\sqrt{2}}|111\rangle", v3),
]:
    display(array_to_latex(ket.data, prefix=f"{label} = "))
    latex_code = label + "=" + array_to_latex(ket.data, source=True)
    print(latex_code)
```
