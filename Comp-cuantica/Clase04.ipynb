```python
from qiskit import QuantumCircuit
from qiskit.visualization import array_to_latex
import numpy as np
```

---
### <font color='3facc9'> Creación de un circuito de un cúbit e inicialización del estado </font>

```python
# Crear un circuito cuántico de un cúbit
qc = QuantumCircuit(1)
# Definir el estado inicial a |1>
ket = [0,1]
# Visualizar el ket
display(array_to_latex(ket, prefix = r"\ket{1} = "))
# Aplicar la operación de inicialización al cúbit
qc.initialize(ket, 0)
# Dibujar el circuito
qc.draw(output='mpl')
```

---
### <font color='3facc9'> Creación de un circuito de tres cúbits e inicialización del estado entrelazado </font>

```python
# Crear un circuito cuántico de dos cúbits
qc = QuantumCircuit(3)
# Definir el estado inicial
ket = [1,0,0,0,0,0,0,1]/np.sqrt(2)
# Visualizar el ket
display(array_to_latex(ket, prefix = r"\ket{\Psi} = "))
# Aplicar la operación de inicialización a los cúbits
qc.initialize(ket,[0,1,2])
# Dibujar el circuito
qc.draw(output='mpl')
```
